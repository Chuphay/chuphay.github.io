# Differential Equations & Rutherford Scattering

#### Differential Equations

Differential equations are pretty important. In this blog post I'm going to try to learn how to do them numerically. 

I'm going to be referencing a book called 'Computational Physics' by Tao Pang for this post.

The first, probably most easiest, thing to examine is a mass under an elastic force (i.e. a mass attached to a spring).

Let's write out Newton's equation (Hooke's law) for this:

$$F_{orce} = mass*acceleration = -kx$$

where k is the spring constant. To make our lives easy, let's just say that $k=mass=1$. Then we can rewrite this equation as

$$a = \frac {dv}{dt} = -x $$

where $v = dx/dt$.

Since we are dealing with computers, we will want to discretize these equations. We should think of a differential equation as evolving the position one small step at a time. We will call this small step $\tau = t_{i+1}-t_i $, thus we write

$$x(t_{i+1}) = x(t_i) + \tau v(t_i) \hspace{4mm} \Longrightarrow \hspace{4mm} x_{i+1} = x_i + \tau v_i$$

$$v(t_{i+1}) = v(t_i) + \tau a(t_i) \hspace{4mm} \Longrightarrow \hspace{4mm} v_{i+1} = v_i + \tau a_i$$

and now we implement this algorithm, called the Euler method, below.

```python
def get_x(old_x,velocity,step):
    return float(old_x + velocity*step)
def get_velocity(old_velocity,acceleration,step):
    return float(old_velocity + acceleration*step)
x = [0]
v = [1]
t = [0]
step = pi/50
for i in range(100): 
    t.append(t[i]+step)
    x.append(get_x(x[i],v[i],step))
    v.append(get_velocity(v[i],-x[i],step))
             
plot(t,x,label ="Euler's method"),plot(t,sin(t),'k',label='Exact solution')
legend()
show()
```

This is okay, but let's see what happens over a longer period of time.

```python
x,v,t = [0],[1],[0]
for i in range(1000): 
    t.append(t[i]+step)
    x.append(get_x(x[i],v[i],step))
    v.append(get_velocity(v[i],-x[i],step))
             
plot(t,x,label ="Euler's method"),plot(t,sin(t),'k',label='Exact solution')
legend(loc='best')
show()
```

Yikes! That's really bad.

Apparently, the most common algorithm to improve the accuracy of a numerical differential equation is the so-called Runge-Kutta method.

#### The Runge-Kutta method

The Runge-Kutta method is derived from the Taylor expansion of $\bf y$ and their derivatives, defined by 

$$\frac {d {\bf y}}{dt} = {\bf g} ({\bf y} , t) $$

where ${\bf y} = (y_1, y_2, ... , y_l)$ and $ {\bf g} ({\bf y}, t) = [g_1({\bf y}, t), g_2({\bf y},t)...g_l({\bf y},t)]$

We expand $y(t+\tau)$ around $\tau$

$$y(t+ \tau) = y + \tau y' + \frac {\tau^2}{2} y'' + \frac {\tau^3}{3!}y'''+...$$

$$= y + \tau g + \frac {\tau^2}{2}(g_t +gg_y)+\frac {\tau^3}{3!}(g_{tt} + 2gg_{ty} + g^2g_{yy} +gg_y^2 +g_tg_y)+...  $$

Lets work out $y''$ for clarity:

$$ \frac {d^2y}{dt^2} = \frac {d}{dt} g(y,t) = \frac {\partial g}{\partial t} + \frac {\partial g}{\partial y} \frac {dy}{dt}=g_t+g_yg$$

We can rewrite the Taylor expansion as

$$y(t + \tau) = y(t) + \alpha_1c_1 + \alpha_2c_2 +...+\alpha_mc_m+...$$

where 

$$c_1 = \tau g(y,t)$$
$$c_2 = \tau g(y+v_{21}c_1,t+v_{21}\tau)$$
$$c_3 = \tau g(y+v_{31}c_1+v_{32}c_2,t+v_{31}\tau++v_{32}\tau)$$
$$.$$
$$.$$
$$.$$
$$c_m = \tau g \left ( y + \sum_{i=1}^{m-1}v_{mi}c_i, t + \tau \sum_{i=1}^{m-1}v_{mi} \right )$$

where $\alpha$ and $v$ are to be determined and $m$ is the level of accuracy desired. One should note that $m=1$ gives the Euler algorithm we were working with above. Let's explicitly work out the case when $m=2$.

We will truncate our Taylor expansions above, and keep only terms that are $O (\tau^2)$:

$$y(t+\tau) = y + \tau g + \frac {\tau^2}{2} (g_t +gg_y)$$

$$y(t+\tau) = y + \tau g + \alpha_1 c_1 + \alpha_2 c_2$$

where

$$c_1 = \tau g(y,t)$$

$$c_2 = \tau g(y+v_{21}c_1,t+v_{21}\tau)$$

Let us now per form a Taylor expansion of $c_2$, keeping terms only up to $O(\tau^2)$:

$c_2 = \tau \left [ g + v_{21} \tau \frac {dg}{dt} \right ]$ where $\frac {dg}{dt} = g_t +gg_y$

Now, we simply gather our terms together to get

$y(t+\tau) = y(t) +(\alpha_1 + \alpha_2) \tau g + \alpha_2 \tau^2 v_{21} (g_t + gg_y)$

Cleary then 

$\alpha_1 + \alpha_2 = 1$  and  $\alpha_2 v_{21} = \frac 1 2$

As we can see, we have two equations but three unknowns. This gives us freedom in choosing our unknowns. A common choice that will also yield a physically intuitive picture is made by setting $\alpha_1 = \alpha_2 = 1/2$ and $v_{21} = 1$. This yields

$$y(t+\tau) = y(t) + \frac \tau 2 \left ( g(y,t) + g(y+\tau g,t +\tau) \right)$$

We can pause here to examine the meaning of this equation. In comparing this equation to the Euler's method above, we can see that we now have an average over two values. One is the slope of the function at the initial point $(y,t)$ and the second is the slope at the final point $(y+\tau g,t +\tau)$. Thus we are calculating the average slope over the two endpoints. 


While $m=2$ certainly gives us a better algorithm than the Euler method, the most popular method is to set $m=4$. I will not go over the details of the derivation and instead just give the final result: 

\begin{align}
{\bf y}_{i+1} &= {\bf y}_i + \frac{1}{6} \left({\bf c}_1 + 2{\bf c}_2 + 2{\bf c}_3 + {\bf c}_4 \right)\\
\end{align}

with

$${\bf c}_1 = \tau{\bf g}({\bf y}_i,t_i)$$

$${\bf c}_2 = \tau{\bf g}({\bf y}_i + \frac{{\bf c}_1}{2}, t_i +  \frac{\tau}{2})$$

$${\bf c}_3 = \tau{\bf g}({\bf y}_i + \frac{{\bf c}_2}{2} , t_i+   \frac{\tau}{2})$$

$${\bf c}_4 = \tau{\bf g}({\bf y}_i + {\bf c}_3, t_i + \tau)$$



Let's try to implement this algorithm

```python
def run_kutta(f,initial,tau):
    c_1 = tau*f(initial)
    c_2 = tau*f(initial+c_1/2.0)
    c_3 = tau*f(initial+c_2/2.0)
    c_4 = tau*f(initial+c_3)
    return initial + (c_1+2*c_2+2*c_3+c_4)/6.0
    
f = lambda y: array([y[1],-y[0]])
  
data = [[0,1]]
for i in range(628):
    x = run_kutta(f, data[i], 0.1)
    data.append(x)
data = array(data)

t  = linspace(0, 20*pi, 629)
plot(t,data[:,1]), plot(t,cos(t))
show()
```

Wow! This is so much better. There are still improvements that can be made to this algorithm. But just as in the previous post, I will instead rely on the algorithms provided by scipy. These algorithms are coded by professional programmers and have been highly optimized.

```python
from scipy.integrate import odeint

f = lambda y,t: [y[1],-y[0]]
t  = linspace(0, 24*pi, 200)   
x = odeint(f, [0,1], t)[:,0]

plot(t,x,label ='numerical'),plot(t,sin(t),'k',label='analytical')
legend(loc='best')
show()
```

And, as expected, this gives a fine result.

### Rutherford Scattering

Now that we have our tools ready, let's use them to do some actual simulations.

Rutherford scattering is an important experiment both historically and theoretically. In this simulation, we should imagine that we are sending an alpha particle at a heavy nucleus. The interaction will be governed by the Coulomb force.

```python
from scipy.integrate import odeint

def f(y,t): 
    x_vel = y[2]/((y[2]**2+y[3]**2)**1.5)
    y_vel = y[3]/((y[2]**2+y[3]**2)**1.5)
    x_pos = y[0]
    y_pos = y[1]
    return [x_vel,y_vel,x_pos,y_pos]

t  = linspace(0, 30, 400) 
data = []
for i in range(41):
    
    soln = odeint(f, [1,0,-5,10-i/2.0], t)
    if soln[-1][2] >= 3:
        data.append([10-i/2.0,soln[-1][2],soln[-1][3]])
    plot(soln[:,2],soln[:,3])
plot(0,0,'k.')
xlim(-5,5)
ylim(-7,7)
show()
```

This looks beautiful and is pretty much exactly what I expected to have.

Now what we are trying to do is find the differential cross section. The differential cross section $\sigma (\theta)$ is the probability of a particle being found in the solid angle element $d\Omega = 2\pi\sin\theta d\theta$ at the deflection angle $\theta$. Or in other words,


$$\frac {d\sigma}{d\Omega} = \frac {N_s}{N_0 \cdot \Delta \Omega}$$

where:

$N_s = $ the number of detected particles at the deflection angle $\theta$. 

$N_0 =$ the total number of particles 

$\Delta \Omega = $ the solid angle of the detector. This is here because real detectors will not have infinitesimal size.

Analytically, we can solve this equation for the Coulomb potential. The result is


$\frac {d\sigma}{d\Omega}  = \frac 1 {16E^2\sin^4(\frac \theta 2)}$ where E is the energy of the incoming particle.


Let's try and take a look at what this looks like.


```python
def sigma(theta, energy=1):
    denom = (16*energy**2)*(sin(theta/2.0))**4
    return 1/denom

x = linspace(0.1,2,100)
plot(x,sigma(x))
show()
```

As we can see, as $\theta$ gets close to zero, the differential cross section becomes very large. And as $\theta$ goes away from zero, the cross section drops off quickly.

But looking at my graph above of Coulomb scattering, we seem to see the opposite effect. When $\theta$ is close to zero, zero particles go through, and as $\theta$ increases more and more particles are scattered.

I'll admit, I was very, very puzzled about this for many days. I finally started looking at websites that explained the actual experiment and I realized how overly idealized my simulation was. In the experiments, often the beam of alpha particles was coming through a gate that was 5mm or even 10mm wide! That's huge!

With this in mind, I redesigned my experiment to be much less precise:


```python
from scipy.integrate import odeint

def f(y,t): 
    x_vel = y[2]/((y[2]**2+y[3]**2)**1.5)
    y_vel = y[3]/((y[2]**2+y[3]**2)**1.5)
    x_pos = y[0]
    y_pos = y[1]
    return [x_vel,y_vel,x_pos,y_pos]



t  = linspace(0, 10, 400) 
data = []
for j in range(5):
    for i in range(1000):
        noise = randn()
        soln = odeint(f, [1,0,-5,50-i/10.0+noise], t)
        data.append([50-i/10.0+noise,soln[-1][2],soln[-1][3]])
        if j == 0:
            plot(soln[:,2],soln[:,3])
plot(0,0,'k.')
xlim(-5,5)
show()
```

As usual, the best way to deal with large amounts of data will be to use Pandas' DataFrame.

```python
from pandas import DataFrame
df = DataFrame(data,columns =['b','x','y'])
df['theta rad']=arctan((df['y']-df['b'])/df['x'])
df['theta deg']=abs(360/(2*pi)*df['theta rad'])
df['theta deg'].hist(bins =100,histtype = 'step')
xlim(0,30)
show()
```

I've plotted the experimental differential cross section and as we can see it is very close to what we expected analytically.
