Next we set up some global variables:

```python
speed = 10.0
amount_of_crazy = 0.2
seed(123) #np.random.seed(123)
wall = 10.0
x = array([0.0,wall]) #np.array([0.0,10.0]) 
```

Here we use a numpy.array for the first time. I don't really get these completely yet, but I'll try to throw them into the code where I can to get some practice with them.

Now we are ready to collect our data. Along the way, we will also plot the first 20 shots

```python
data = []
for i in range(1000):
    theta = amount_of_crazy*randn() #np.random.randn()
    velocity = speed*array([cos(theta),sin(theta)])*(1+amount_of_crazy*randn())

    time = wall/velocity[0]
    y = array([0.0,velocity[1]*time])
    if i<20:
        plot(x,y) #plt.plot(x,y)
    
    #the only two pieces of interesting data are the time and the final y-position
    #so we will add them to our data set
    data.append([time,y[1]])

xlabel('x-axis') #plt.xlabel('x-axis') 
ylabel('y-axis') #plt.ylabel('y-axis')
show() #plt.show() ...actually, we don't even need this line
```

Great! Now we have some data! Let's take a look at what we got.

What I want to do now, is try to fit this data to a model. The y-displacement is easy, it looks just like a bell curve, so let's try and fit this to normal distribution.

```python
z,x=np.histogram(df['y-displacement'],bins=100)
x = x[:z.size] #here we call size on the np.array instead of length
plot(x,z)

#I found about curve_fit from the folowing tutorial
#http://nbviewer.ipython.org/gist/keflavich/4042018
from scipy.optimize import curve_fit
normfunc = lambda x,a,b,c,d: a*np.exp(-((x-b)/c)**2)+d #lambda function... don't really understand yet
fitpars, covmat = curve_fit(normfunc,x,z) #check out either the tutorial or the documentation on using this
plot(x, normfunc(x, *fitpars), 'r-')
show()
```

So that looks like a very nice fit.

Next I want to see if I can model the time. This looks like a skewed normal distribution. After checking on stack overflow, this is what I got

```python
#http://stackoverflow.com/questions/5884768/skew-normal-distribution-in-scipy
#http://en.wikipedia.org/wiki/Skew_normal_distribution

from scipy.special import erf

def pdf(x):
    return 1/sqrt(2*pi) * exp(-x**2/2)

def cdf(x):
    return (1 + erf(x/sqrt(2))) / 2

def skew(x,e=0,w=1,a=0,p=1):
    t = (x-e) / p
    return 2 / w * pdf(t) * cdf(a*t)


df['time'].hist(bins = 100, histtype = 'step')
z,x=np.histogram(df['time'],bins=100)
x = x[:z.size]
fitpars, covmat = curve_fit(skew,x,z)
plot(x, skew(x, *fitpars), 'r-')
show()
```

And this seems like a nice fit as well.

Last, but not least, we can also get some information about the speed at which the ball was travelling at:

```python
df['distance'] = (df['y-displacement']**2+100)**0.5
df['speed'] = df['distance']/df['time']
df['speed'][:40].plot(marker='o')

linear = lambda x,a,b: a*x+b
x = arange(40) #this makes a numpy.array from zero to 39
fitpars, covmat = curve_fit(linear,x,df['speed'][x])
plot(x, linear(x, *fitpars), 'r-')
show()
```
