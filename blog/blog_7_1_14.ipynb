# Percolation and Breadth-First Search

#### Percolation

Percolation is a lot of fun. It's easy to understand yet still possess complex and interesting behavior. 

Rather than trying to explain what percolation is, I think it would be simplest just to show what it looks like.


```python
prob = 0.5
size = 10
figsize(12,8)
seed(123)

plot_data = []
data = {key: [] for key in range(1,(size)**2+1)}
for i in range(size):
    for j in range(size):
        scatter(i,j,c='b')
        if(j<(size-1) and rand()>prob):
            plot_data.append([j,j+1])
            plot_data.append([i,i])
            plot_data.append('k')
            data[i*size+j+1].append(i*size+j+2)
            data[i*size+j+2].append(i*size+j+1)
        if(i<(size-1) and rand()>prob):    
            plot_data.append([j,j])
            plot_data.append([i,i+1])
            plot_data.append('k')
            data[i*size+j+1].append((i+1)*size+j+1)
            data[(i+1)*size+j+1].append(i*size+j+1)
plot_data = array(plot_data)
plot(*plot_data);
```

Now that we have a visual representation of the model it is extremely easy to explain what percolation is. We lay out a square grid with each intersection represented by a node. Between each node, there is a possibilty of a bond. For our example in this blog post we have chosen the probality of a bond to equal the probability of there not being a bond. If there is a bond, we then say that the two nodes are connected.

One should also see that various different models could have been employed. For example, instead of being a square lattice, one could have a tree-type system. Or instead of having bonds being either on or off, there could be degrees of connectivity. 

However, bond percolation on a square lattice has some particularly simple and transparent properties, so we will use it as our canonical example. 

One of the immediate concerns, and the problem that will occupy us in this blog post is simply mapping the grid. What we want to be able to do is start from any node and then search and find every other node that is connected to that node. To do that we will employ breadth first search.

#### Breadth-First Search

Our primary goal is to search every node that is connected, however distantly, to the initial node. One thing that should be obvious from this statement is that if a node is not in any way connected to the initial node from which you start searching, then that node will not be reachable. 

Breadth-first search searches in layers. We will first give our program an intial node to start searching from. The algorithm will first find all the nodes that are connected to that node. Then it will go through every node that it has found and do the same thing again on the next node. In this way, it is guaranteed that all nodes are searched and nodes that are closest to the parent node are searched first.

#### Queue Data Structure

In order to implement this algorithm we will have to first implement a queue data structure. The queue will keep track and make sure that we really are first searching the oldest modes. It does this by doing what is called "first-in first-out", i.e., exactly like a queue at a supermarket is supposed to behave.

We will build a class to implement this data structure. The class will only have two methods: 'add' a new object to the back of the queue and 'serve' the object that is at the front of the queue.

```python
class queue(object):
    def __init__(self):
        self._length = 0
        self._max = 100
        self._position = 0
        self._array = [0 for i in range(self._max)]
    def add(self,value):
        if(self._length+3>self._max):
            raise RuntimeError("Sorry, ran out of memory")
        else:
            self._array[(self._length+self._position)%self._max] = value;
            self._length += 1
    def serve(self):       
        if(self._length>0):
            out = self._array[self._position]
            self._length -= 1
            if(self._position<self._max-1):
                self._position += 1
            else: 
                self.position = 0
            return out    
        else:
            raise IndexError("nothing is in the list :( ")
     
```

Now that the queue has been implented a depth-first search is essentially trivial to implement. When given a starting node and a map. We must find all of the nodes that are connected to the initial node and add them to the queue. Then we simply go to each node in the queue and find all nodes connected to that node and add them to the queue (remember, it will be adding to the back of the queue). And then simply iterate until all nodes have been discovered.

```python
figsize(12,8)

for i in range(size):
    for j in range(size):
        scatter(i,j,c='b')
      
def BFS(data, node):
    visited = {key: False for key in range(1,(size)**2+1)}
    visited[node] = True
    q = queue()
    for i in range(len(data[node])):
        q.add(data[node][i])
    while(q._length>0):
        new_node = q.serve()
        if(visited[new_node]==False):
            visited[new_node] = True
            for i in range(len(data[new_node])):
                q.add(data[new_node][i])
                plot([new_node%size-1,data[new_node][i]%size-1],[new_node/size,data[new_node][i]/size],'b',linewidth=1.5)
    return visited            

plot(*plot_data);
BFS(data,1);
```

Here I have started from the first node and mapped everything connected to it.
