## Depth-First Search & the Critical Point

In my last [post](/blog/7_1_14.html) I introduced percolation and I introduced breadth-first search to search through connected components.

In this blog post, I want to extend that discussion in several ways.

First of all, let's use last post's machinery and get a graph set up:

```python
size = 25
figsize(12,8)
seed(123)

def make_graph(probability,size, plot_on = False):
    prob = 1 - probability
    plot_data = []
    data = {key: [] for key in range(1,(size)**2+1)}
    for i in range(size):
        for j in range(size):
            if plot_on:
                scatter(i,j,c='b')
            if(j<(size-1) and rand()>prob):
                data[i*size+j+1].append(i*size+j+2)
                data[i*size+j+2].append(i*size+j+1)
                if plot_on:
                    plot_data.append([j,j+1])
                    plot_data.append([i,i])
                    plot_data.append('k')
            if(i<(size-1) and rand()>prob):   
                data[i*size+j+1].append((i+1)*size+j+1)
                data[(i+1)*size+j+1].append(i*size+j+1)
                if plot_on:
                    plot_data.append([j,j])
                    plot_data.append([i,i+1])
                    plot_data.append('k')
    if(plot_on):
        plot_data = array(plot_data)
        plot(*plot_data) 
    return data  

data = make_graph(0.45,size,True)
```

In this post, instead of using a breadth-first search, I want to introduce depth-first search. You hopefully recall that in a breadth-first search, we use a data structure dubbed a queue. A queue supports a first-in first-out data hierarchy. This means that as you add in new nodes to search, the algorithm will actually first search the oldest nodes before moving on to any new nodes. Depth-first search is the complete opposite of this approach. Whenever a new node is introduced, that node is explored. Only when it has reached a node that has no new explorable nodes will the algorithm back-track and explore older nodes.

One can implement this in two ways. One can design a data structure called a stack which supports a last-in first-out data hierarchy, or more simply just implement it using recursion, which is what I have done:

```python
def DFS(data, node, size, stop = False, plot_on = True):
    """stop is here in case you want to want to stop the search
    as soon as you have reached the bottom of the graph.
    This is good to check when percolating down.
    visited[0] will contain that information
    """
    visited = {key: False for key in range(0,(size)**2+1)}

    def recurse(data, node):
        if plot_on:
            scatter((node-1)%size,(node-1)/size, c='r',s=30,lw=0)
        visited[node] = True
        if ((node-1)/size == 0):
            if stop:
                visited[0] = True
                return visited
        for i in range(len(data[node])):
            if (visited[0] == True):
                break
            new_node = data[node][i]
            if plot_on:
                plot([(node-1)%size,(data[node][i]-1)%size],[(node-1)/size,(data[node][i]-1)/size],'r',linewidth=1.5)
            if(visited[new_node]==False):
                recurse(data, new_node)
        
        return visited        
    return recurse(data, node)            
```

Now, I think we are finally in position to actually talk about where percolation got the name percolation. 

Imagine either a tightly packed cup of espresso coffee grinds or perhaps a highly porous rock (like pumice stone) and a layer of water over the top of it. The water would slowly seep into the espresso grinds in a semi-haphazard way. In fact, you can imagine that if you pack the espresso grinds too tightly, only a very small amount of water will penetrate the top most surface and none will ever actually make it to the bottom.

We are very interested in this process. 

In the next bit of code, I have ran through every node on the top of the graph, and performed a depth-first search for all connected nodes. Check out the results:

```python
for i in range(size):
    for j in range(size):
        scatter(i,j,c='b')
        
def percolate_down(data, size, plot_on = True):
    out = False
    visited = {key: False for key in range(1,(size)**2+1)}
    for i in range(size):
        node = size*size - i
        if (visited[node] == False):
            new_visited = DFS(data, node, size, True, plot_on)
            visited = {x: visited.get(x) + new_visited.get(x) for x in visited}
            if (new_visited[0] == True):
                out = True
                break
    return out

percolate_down(data,size);
```

As you can see, with a 45% probability of connecting two nodes the 'water' does not reach the bottom.

However, what would happen if there was a 55% chance of connection between two nodes?

Before looking at the graph, look back up at my DFS code and notice that I have included some code that will halt the depth-first search if it reaches the bottom of the graph.

```python
seed(14)
for i in range(size):
    for j in range(size):
        scatter(i,j,c='b')
data = make_graph(0.55,size)
percolate_down(data,size);
```

Here we see that with a 55% probabibilty of connection, the 'water' reaches the bottom of the graph very rapidly.

This immediately begs a question. Is there some critical probabilty where on the one side the 'water' never reaches the bottom and on the other side the 'water' always reaches the bottom? On small graphs, perhaps the answer can be very dependent on the random flips of a coin. But one can instead think of extremely large graphs, perhaps even infinite.

Before getting to the answer of this question, I want to put some code down that will help us to make many simulations. A very cool feature of Python is that we can make functions that modify or 'decorate' other functions. We want to make a decorator that will repeat a simulation many times so that we can get a good average of what the underlying simulation is revealing.

I have heavily modified a bit of [code](http://www.artima.com/weblogs/viewpost.jsp?thread=240845) to do just that:

```python
def repeat(times):
    """For example:   
    
    @repeat(1000000)
    def getRand():
        return np.random.rand(3)
    print getRand()
    
    #should print something like array([0.5,0.5,0.5])
    """
    def wrap(f):
        def wrapped_f(*args):
            output = f(*args)
            for i in range(times-1):
                output += f(*args)  
            return output/float(times)
        return wrapped_f
    return wrap
```

Now we can run our simulation many times to get a good average behavior. However, because of the computational expense, I have kept the numbers rather small.

```python
figsize(7,5)

@repeat(100)
def reached_bottom():
    out = []
    for i in range(100):    
        data = make_graph(i/100.0,size)
        if percolate_down(data, size, False):
            out.append(1)
        else:
            out.append(0)
    out = array(out)    
    return out
z = reached_bottom()
plot(z);
```

Here we can see that at 40% there is basically zero chance of the 'water' reaching the bottom, and at 60% the 'water' reaches the bottom almost surely. And in fact, the bigger the graph the more those numbers will narrow. 

We should also notice that this shape looks an awful lot like a Sigmoid function. Let's use the curve fitting functionality of Scipy to check just how close:

```python
from scipy.optimize import curve_fit
x = linspace(0,100,100, endpoint=False)

sigmoid = lambda x,a,b: 1/(1+exp(-(x+b)/a))
fitpars, covmat = curve_fit(sigmoid,x,z) 
plot(z);
plot(x, sigmoid(x, *fitpars), 'r-');
```

While on the limited sized graphs that we are using, it is impossible to turn the Sigmoid function into a true step function, there is however a mathematical proof that indeed at 50% there is a critical point.
