## The Logistic Equation & Euler's Method

Recently I re-read "Chaos" by James Gleick. I had read this book once before but it was before I had learned how to code. Now that I can actually program the things that he was talking about in the book, I feel really excited to give it a try.

One of the first topics that really jumped out at me was the logistic equation. This is a fairly simple differential equation that I have had previous dealings with and I could hardly believe that it had so many 'magical' properties.

In a previous [post](/blog/4_1_14.html) we discussed differential equations and two algorithms to solve them. The first algorithm we discussed was Euler's method. Before geting to the logistic equation proper, it will be beneficial to review Euler's method. 

#### Euler's Method

The general equation that we are interested in solving has the following form: 

$$\frac {dx}{dt} = f(x,t) $$

by the definition of a derivative, this is equal to

$$ \lim_{t_{i+1} \rightarrow t_i} \frac {x(t_{i+1}) - x(t_i)}{t_{i+1}-t_i} = f(x(t_i),t_i)$$  

setting $\tau = t_{i+1}-t_i $, and rearranging the terms leaves us with Euler's method:

$$ x_{i+1} = x_i + \tau f(x_i,t_i)$$

I'll be honest. In all of my previous experience in math, I had taken it for granted that Euler's method would give approximately the same solution as the exact solution. But as we will see, the solutions actually behave radically different.

#### The Logistic Equation

The logistic equation is really, quite simple and unassuming. Here it is:

$$\frac {dx}{dt} = x(1-x)$$


As is well known, the solution to this differential equation is

$$x = \frac 1 {1+e^{-t}}$$

which we can easily check

$$\frac {dx}{dt}= \frac {-1}{(1+e^{-t})^2} (-e^{-t}) = \frac 1 {1+e^{-t}} \frac {e^{-t}}{1+e^{-t}} = \frac 1 {1+e^{-t}} \left(1 - \frac 1 {1+e^{-t}} \right) = x(1-x)$$

The first thing to notice is that the solution is very well-behaved. It is bounded and continuous, and defined for all values of $t$. It also worth noting that the logistic equation has two trivial solutions. When $x$ is either 0 or 1, $x$ will remain constant. 

Let's make a graph of the solution to see what it looks like.

```python
t = linspace(-5,5,100)
plot(t,1/(1+exp(-t)));
```

We've actually already already met this equation in my last blog post, although there I called it a Sigmoid function. Its other name is the logistic function because it solves the logistic equation.

Looking at the graph, it is quite clear that this equation has only one lerge-term solution. As $t$ grows large, the function approaches 1. There really does not seem to be anything mysterious about this at all.

However, we will see that we are going to get much different behavior when we use Euler's method.

#### Discretizing the Logistic Equation

We write the logistic equation again:

$$\frac {dx}{dt} = x(1-x)$$

and apply Euler's method to get:

$$ x_{i+1} = x_i + \tau x_i(1-x_i)$$

Writing the code to execute this method is trivial.

```python
get_next = lambda x, step: x + step*x*(1-x)
```

Now, let's take a look at what we get. 

```python
z = [0.1]
for i in range(100): 
    z.append(get_next(z[i],0.1))
plot(z); 
```

Well... this is looking like a good recommendation for Euler's method.

However, it is prudent to note that we chose an intitial value of 0.1. As we already noticed, if we start with a value of 0, then the solution remains static. But even more interesting is that the Euler method does not realize solutions for negative starting values.

Let us take a look at what happens when we increase the step size:

```python
z = [0.1]
for i in range(100): 
    z.append(get_next(z[i],1.9))
plot(z); 
```

Here we can see that Euler's method slightly over-shoots the target, and then briefly oscilates around 1.0, before decaying to the final solution.

Let us again increase the step size, but only by a small amount:

```python
z = [0.1]
for i in range(100): 
    z.append(get_next(z[i],2.1))
plot(z); 
```

Amazingly, only by increasing the step size a little, we see that the solution never decays, and continues to oscillate.

Increasing the step size again yields stunning results:

```python
z = [0.1]
for i in range(100): 
    z.append(get_next(z[i],3.0))
plot(z); 
```

This graph shows what is commonly called 'chaos'.

To get a sense of what is going on here, we want to plot the value of the process after a certain number of steps versus the step size:

```python
figsize(15,10)

data = []
r = linspace(1,3.5,10000)
for step in r:
    z = [randn()]
    for i in range(100): 
        z.append(get_next(z[i],step))
    
    data.append(z[-1])


z = array(z) 
scatter(r,data
show()
```
