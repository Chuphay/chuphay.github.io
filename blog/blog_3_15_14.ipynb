# Fourier Transforms

Fourier transforms are pretty cool. Today we are going to discuss two import aspects of them. 

First of all, they are useful in modeling periodic functions.

If we have a function $f(x)$ then the fourier transform is defined as

$$A_{k} = \int_{-\infty}^{\infty} f(x) e^{-ikx} dx$$

From this definition, we can build an approximation of $f(x)$ in the following way

$$f(x) = A_{0} + A_{1}e^{ix} + A_{2}e^{2ix} +  A_{3}e^{3ix} + ...$$

Obviously the more terms included in this expansion, the better the model will be.

The first thing I want to model is a sawtooth wave:

```python
L = 2
n = 1024
t = linspace(0,L,n, endpoint=False)
figsize(12,4)
y = lambda x,n=0: (x-n)
plot(t[:n/2],y(t[:n/2]),'k',t[n/2:],y(t[n/2:],t[n/2]),'k')
plot([t[n/2-1],t[n/2]],[1,0],'k',[t[n-2],t[n-1]],[1,0],'k',[t[0],t[1]],[1,0],'k')  
xlim(0,2+0.005)
show()
```

But here it is anyways:

```python
from scipy.integrate import quad #this is just standard integration
def fourier(function,k,a=0,b=2*pi):
    """hmmm... assuming the range is from 0 to 2*pi.
    function should be some kind of lambda like function
    k is the kth coefficient, which is supposedly what will be returned to you 
    """
    fn = lambda x: function(x)*exp(1j*k*x*2*pi/(b-a))
    real_part = lambda x: real(fn(x)) #sp.real()
    imag_part = lambda x: imag(fn(x))
    return 1/(b-a)*quad(real_part,a,b)[0]+1j/(b-a)*quad(imag_part,a,b)[0]
    
```

Now let's use this fourier transform algorithm to model the sawtooth wave.

I will plot the first 6 contributions and the first 49 contributions to our model.

```python
comp = 0
for i in range(1,50):
    y = lambda x: x
    comp+=real(fourier(y,i,0,2))*cos(2*pi*t*i)+imag(fourier(y,i,0,2))*sin(2*pi*t*i)
    if i==6 or i==49:
        plot(t,comp)   
```

Summing the first 6 contributions already gives us a pretty good approxiamtion, and the first 49 gives us a really nice approximation. One should notice that at the boundaries there is some pretty wild fluctuations... this is called the Gibbs phenomenon, you can look it up in Wikipedia.

Like I said, my algorithm is broken... luckily the scipy stack has fourier transform available for us to use.

These are tools built by professionals.

Let's try and use them and reproduce the results we have above.

```python
from scipy.fftpack import fft
fast_f = fft(t)
comp = 0
for i in range(1,50):
    comp+=real(fast_f[i]*exp(2j*pi*t*i)/1000)
    if i==6 or i==49:
        plot(t,comp)  
```

The second aspect of Fourier transforms that I want to look at today is that the Fourier transform allows us to look at our function in frequency space instead of in position space.

This is really cool, because it gives us a second, alternative way to look at our data and this may give us more insight into what we are examining.

Instead of working with the sawtooth wave, let us work with a gaussian (normal) curve. A gaussian curve has the important property that the fourier transform of a gaussian is still a gaussian.  

Let's check this out.

First, let's do it with my equations:

```python
L = 10
n = 1024
x = linspace(-L/2,L/2,n, endpoint=False)

gauss = lambda x: exp(-(x**2))
figsize(12,4)
subplot(121), plot(x,gauss(x),'k'), xlabel('x-space')

myK = linspace(-4*pi,4*pi,n, endpoint=False)
coef = []
for i in x:    
    coef.append(fourier(gauss,i))
coef = array(coef)
subplot(122), plot(myK,coef,'m'), xlabel('k-space')
show()
```

Ok. This worked exactly as expected

let's now do this with scipy's fft:

```python
L = 10
n = 64
x = linspace(-L/2,L/2,n, endpoint=False)

gauss = lambda x: exp(-x**2)

gauss_t = fft(gauss(x))
figsize(12,4)
subplot(121), plot(x,gauss(x),'k'), xlabel('x-space')

myK = linspace(-4*pi,4*pi,n, endpoint=False)
coef = []
for i in x:    
    coef.append(fourier(gauss,i))
coef = array(coef)
subplot(122), plot(myK,gauss_t,'m'), xlabel('k-space'), ylim(-10,10)
show()
```

So that looks really wrong...

What happened?

It actually turns out that this is just the way the fft algorithm works.

To get it back to normal, we use two new functions:

```python
#http://stackoverflow.com/questions/9456037/scipy-numpy-fft-frequency-analysis
from scipy.fftpack import fftshift, fftfreq
L = 40
n = 1024
x = linspace(-L/2,L/2,n, endpoint = False)
u = exp(-x**2)
gauss = lambda x: exp(-(x**2)/2)
ut = fft(gauss(x))
k = 2*pi/L*fftfreq(n,1.0/n) #one can also use this: 2*pi/L*append(arange(n/2),arange(-n/2, 0))
plot(fftshift(k),abs(fftshift(ut)),'m'), xlim(-5,5)
show()
```

So let's go ahead and learn to program the FFT algorithm ourselves.

From wikipedia: http://en.wikipedia.org/wiki/Fast_Fourier_transform and http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm

The DFT (discrete fourier transform) is defined by the formula

$$X_{k} = \sum\limits_{n=0}^{N-1}x_{n}e^{-2\pi ikn/N} \hspace{12 mm} k = 0, 1, ..., N-1$$ 

We can then seperate the odd and even terms

$$X_{k} = \sum\limits_{n=0}^{N/2-1}x_{2n}e^{-2\pi ik(2n)/N} + \sum\limits_{n=0}^{N/2-1}x_{2n+1}e^{-2\pi ik(2n+1)/N}$$ 

$$ = x_{n} + w_{n}y_{n}$$

where

$$ x_{n} = \sum\limits_{n=0}^{N/2-1}x_{2n}e^{-2\pi ikn/(N/2)} ; \hspace{2 mm} y_{n} = \sum\limits_{n=0}^{N/2-1}x_{2n+1}e^{-2\pi ikn/(N/2)}; \hspace{2 mm} and \hspace{2 mm} w_{n} = e^{-2\pi ik/N}$$

As far as I can tell, this so far doesn't actually give us any speed boost. What does give us a speed boost is the observation that we actually only need to calculate the first half of these terms because 

$$ X_{k} = x_{n} + w_{n}y_{n} \hspace{12 mm} for \hspace{2 mm} n < N/2 \hspace{2 mm} and$$
$$ X_{k+N/2} = x_{n} - w_{n}y_{n} \hspace{12 mm} for \hspace{2 mm} n < N/2 $$

I believe this is where we get the big speed up.

Here is my code:

```python
def myFFT(l):
    L =len(l)
    if L == 1:
        return l
    elif L%2 == 0:
        out = [0 for i in xrange(L)]
        
        for i in xrange(L/2):
            A = myFFT(l[::2])
            B = myFFT(l[1::2])
            out[i] = (A[i] + B[i]*exp(-2j*pi*i/L))
            out[i+L/2] = (A[i] - B[i]*exp(-2j*pi*i/L))
        return out
    else:
        raise IndexError("powers of 2 please")
    
        

L = 10
n = 64
x = linspace(-L/2,L/2,n, endpoint = False)
u = exp(-x**2)
subplot(121), plot(u,'k'), xlabel('x-space')
ut = myFFT(u)
subplot(122), plot(ut,'m'), xlabel('k-space'), ylim(-10,10)

show()
```
