# Random Numbers & Random Walks

#### Uniform Random Numbers

Random numbers in a computer program are obviously not really random. While you can think of ways to make true random numbers with hardware, there is no way to do this with only software. However, this is not as bad as it seems, for two reasons. First of all, there are some really nice algorithms out there that generate, to a good approximation, random numbers. We call these algorithms 'pseudo-random number generators' or PRNGs. The second reason is perhaps not as obvious at first. We can use the fact that the random numbers are not truly random to our benefit. The way we do this is by setting an intitial seed such that the random generator will generate the same numbers everytme. Why would we want to do this? The reason is this allows us to repeat the exact same simulation many times, meaning that any simulation we make is reproducible, which is very important in scientific calculations. 

The most well known PRNG is the linear congruential generator, which uses

$X_{n+1} = (a X_n + b)\, \textrm{mod}\, m$

to generate a new number $X_{n+1}$ from the old number $X_{n}$. The numbers $a$, $b$, $m$ are known as <em>magic numbers</em>. A particularly good choice is $a=7^5=16807, b = 0,$ and $c=2^{31}-1=2147483647$. You can check out [wikipedia](http://en.wikipedia.org/wiki/Random_Number_Generator) for more information.

This algorithm will generate what is known as a 'uniform distribution' $\in [0,1]$ meaning that it will make random numbers evenly distributed between zero and one. 

I implement this algorithm below:

```python
from datetime import datetime
now = datetime.now()
seed = now.year + 70*(now.month + 12*(now.day + 31*(now.hour + 23*(now.minute +59*now.second))))

def ranf(*args):

    if len(args) == 0:
        global seed
        x = seed
    else:
        x = args[0]
    a, c= 16807, 2147483647
    q, r= c/a, c%a
    h, l = x/q, x%q
    t = a*l - r*h
    if t>0:
        seed = t
        return float(t)/c
    else:
        seed = c+t            
        return float(c+t)/c 
```

In this program, we have a global variable called 'seed'. This variable can either be set by hand so that simulations are reproducible, or if it is not set by hand, then the program will generate a seed based on the current time.

Park and Miller (see Tao Pang's 'An Introduction to Computational Physics' for more information) implemented this algorithm in 1988. In order to show that the algorithm is implemented correctly, we can set the initial seed to be 1, and then after 10,000 steps, we should have 1,043,618,065 returned as the value of the seed. 

```python
seed = 1
for i in xrange(10000):
    x = ranf(seed)   
seed == 1043618065
```

Following Tao Pang's book, we can use the numbers generated above to create other types of random-numbers. The most useful distribution for science is probably the Gausssian or normal distribution. To make this distribution, we must first create an exponential distribution.

To make an exponential distribution we use the uniform distribution from above and make use of the following relations:

We first start with the differential equation

$$f(y)dy=dy=p(x)dx = e^{-x}dx$$

whose solution is

$$y(x) - y(0) = 1 - e^{-x}$$

Then we can set $y(0) = 0$ and invert the above equation to give

$$x = -\ln(1-y),$$

this relates the exponential distribution of $x \in [0,\infty)$ to the uniform distribution of $y \in [0,1]$, which is of course the distribution that we already made above. 

```python
def rane(*args):
           
    if len(args) == 0:
        global seed
        x = seed
    else:
        x = args[0]
    return -log(1-ranf(x))   
```

Now we are in position to generate the Gaussian distribution

$$g(x) = \frac 1 {\sqrt {2 \pi}} e^{-x^2/2}$$

We must use both the uniform distribution $f(\phi) = 1$ and the exponential distribution $p(t) = e^{-t}$ to generate the Gaussian distribution. We first must shift the uniform distribution from $\phi \in [0,1]$ to $\phi \in [0,2\pi]$. This is easy, we simply multiply the output of the uniform distribution by $2\pi$. We then attempt to obtain two Gaussian distributions $g(x)$ and $g(y)$ with the following relations:

We first set up the differential equation

$$\frac 1 {2 \pi} f(\phi)d \phi ~p(t) dt = g(x) dx ~g(y) dy $$

which gives

$$e^{-t}dtd \phi = e^{-(x^2 + y^2)/2}dxdy$$

This equation has the solution 

$$x = \sqrt {2t} \cos \phi $$
$$y = \sqrt {2t} \sin \phi $$

Where $t$ is a random number taken from the exponential distribution and $\phi$ is a random number taken from the uniform distribution in the range $[0,2 \pi ]$.

Below, is the implementation of this algorithm:

```python
def rang(*args):
              
    if len(args) == 0:
        global seed
        x = seed
    else:
        x = args[0]
    rdm1 = - log(1-ranf(x))
    rdm2 = 2*pi*ranf(seed)
    return rdm1*cos(rdm2)  
```

We can get a good feel for what these distributions are doing by taking a look at their histograms:

```python
figsize(14,2)

x = [ranf() for i in range(10000)]
x,y = histogram(x, bins =100, range=(-2,3))
subplot(131), plot(y[:len(x)],x/10000.), xlabel('uniform distribution')

y = [rane() for i in range(10000)]
x,y = histogram(y, bins =100, range=(-1,6))
subplot(132), plot(y[:len(x)],x/10000.), xlabel('exponential distribution')

z = [rang() for i in range(10000)]
x,y = histogram(z, bins =100, range= (-3,3))
subplot(133),plot(y[:len(x)],x/10000.), xlabel('normal distribution')

show()
```

### Random Walks

In attempting to simulate the price of a stock in the stock market, we will use a random walk. 

A random walk is a very simple process. Imagine flipping a coin, everytime the coin lands on heads, you take one step forward and one step to the left. Everytime the coin lands on tail, you take one step forward and one step to the right.

This is trivial to implement in code. We simply take samples from a gaussian distribution and every time the sample is positive, we move forward and up, and every time the sample is negative, we move forward and down.

Below, I've implemented a code that does this 10 times

```python
np.random.seed(123)
data = [0]
for i in range(10):
    coin = randn()
    if coin<0:
        coin = -1
    else:
        coin = +1
    data.append(coin+data[i])
print data
plot(data), xlabel('random walk')
show()
```

The most interesting thing about the random walk is how well it is actually able to simulate a stock on the stock market. Below, I have uploaded the stock price of the S&P 500 for the last 5 years, and then I have taken the last 100 data entries and plotted this.

```python
import pandas.io.data as web
data = web.get_data_yahoo('SNP', '4/21/2009', '4/21/2014') #S&P
data = data.drop(['Open','High','Close','Low','Volume'],1)
figsize(5,5)
plot(data['Adj Close'][-100:])
show()
```

We want to compare this with a random walk. I will plot 6 different walks, so that we can get an idea of how good are random approximation is. 

```python
np.random.seed(123)
figsize(13,7)
for j in range(6):
    data = [85]
    for i in range(100):
        coin = randn()
        if coin<0:
            coin = -1
        else:
            coin = +1
        data.append(coin+data[i])
    subplot(3,2,j+1)
    plot(data)
show()
```
