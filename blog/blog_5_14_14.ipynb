# Histograms, Scaled Random Walks, & Brownian Motion

In my last [blog post](/blog/blog_4_21_14) we simulated the price of a stock with the path of a random walk. 

In this blog post we want to try and do better.

The first thing we want to do is look at the daily returns of a stock. This is clearly defined as the price of the stock at the end of the day minus the price at the beginning all divided by the price

For example, if a stock costed 80 dollars in the morning, and at the end of the day is valued at 88 dollars than the change was 8 dollars and the percent change was 10%, or

$(88-80)/80 = 0.1 \rightarrow$ 10% 

Let's again take a look at the S&P, and plot the daily returns

```python
import pandas.io.data as web
from pandas import DataFrame, Series
price = web.get_data_yahoo('SNP', '5/14/2009', '5/14/2014')['Adj Close'] #S&P

daily_returns = (price[1:].values-price[:-1].values)/price[:-1].values

plot(daily_returns[-100:])
show()
```

Pandas already has a way to implement this:

```python
returns = price.pct_change()
map(lambda x: round(x,6), returns.values[1:]) == map(lambda x: round(x,6),  daily_returns)
```

Examining the above graph should hopefully give us a very good picture of what's going on:

Chaos.

But is it really chaos? How many times is the daily rate 600% ?

The answer is of course none. How many times is the rate less than 50%?

Well, it looks like all of the changes are less than 50%, and in fact are less than 10%. So, this isn't really chaos... there is a pattern and we have to go and find it.

What we want to first do is write a program that will check a list of numbers and see if they are within some range of values: 

```python
def number_in_range(rng, lst):
    start,end = rng
    count = 0
    for value in lst:
        if (value<end and value >=start):
            count += 1
    return count        

    
number_in_range([-.1,.1],returns[-100:])    
```

Now that we have this function, we will want to divide our data into several different bins and then plot the result. The number of bins is somewhat arbitrary, but I'll start with 20 bins:

```python
def get_count_per_bins(bins,lst):
    out = [0 for i in range(len(bins)-1)]
    for i in range(len(bins)-1):
        out[i] = number_in_range([bins[i],bins[i+1]],lst)
    return out    

bins = linspace(-0.1,0.1,21)
a = get_count_per_bins(bins,returns)  
plot(bins[:20],a); show()
```

#### Histograms

We see that we,ve managed to turn what looked like chaos before into a fairly well formed gaussian bell-curve. When we count the numbers of data points divided into bins, this is called generating the [histogram](http://en.wikipedia.org/wiki/Histogram) of the data.

Before moving onto making the random walk a better simulation of a the price of a stock, let's see if we can't make a faster histogram algorithm. Let's first check how quick our original algorithm runs.   

```python
test_data = randn(100000)
bins = linspace(-5,5,41)
%timeit get_count_per_bins(bins,test_data) 
```

And now we make a new algorithm that will only loop through the data one time

```python
def quick_histogram(bins,data):
    bin_width = (bins[-1] - bins[0])/(len(bins)-1)
    out_length = len(bins)-1
    out = [0 for i in range(out_length)]
    for value in data:
        bin_number = int(value/bin_width+out_length/2)
        try:
            out[bin_number] += 1
        except IndexError:
            print "Value %s did not fit into a bin" %value
    return out        

%timeit quick_histogram(bins,test_data) 
```

So we've managed to get the speed down by almost a factor of four! But as usual, the scipy stack also has its own implentation, let's see how fast their algorithm runs:

```python
%timeit histogram(test_data,bins = 40)
```

Wow! ... well, that's why we rely on the algorithms already made for us.

#### Histogram of the Random Walk

Now that we have seen what the histogram of the S&P looks like, let's turn to the random walk and see what it's histogram looks like.

```python
seed(123)
data = [0]
for i in range(100):
    coin = randn()
    if coin<0:
        coin = -1
    else:
        coin = +1
    data.append(coin+data[i])
    
figsize(12,5)
data = Series(data)
subplot(121); plot(data)

change = data[1:].values - data[:-1].values
subplot(122); hist(change, bins =20, range = [-1,1])
show()
```

Hopefully this histogram didn't surprise any of us. The change in the value is either going to be negative one or positive one. There are no other options. Well then how can we make the random walk better? 

#### Multiple-realizations of the Random Walk

While one realization of the random walk was rather uninteresting, let's take a look at what happens when we run many realizations of the random walk:

```python
seed(123)
figsize(13,7)
last_value = []
for j in range(1000):
    data = [0]
    for i in range(500):
        coin = randn()
        if coin<0:
            coin = -1
        else:
            coin = +1
        data.append(coin+data[i])
    last_value.append(data[-1]/sqrt(500))    
figsize(5,5)
hist(last_value, bins =20); show()
```

And this looks much more like what we want the daily return to look like. So then, the solution is that between each interval of time, we should run a random walk, and this should give us a much better looking simulation. 

#### Some Probabilty Theory

Before implementing this, let's take a look at the math to get a more complete picture of what we are lookin at. As you might guess, since the random walk is essentially akin to flipping a coin, we will need to use a little bit of probability.

The three quantites we need to look at are the probablitity of an event, $\mathbb{P}$ , the expectation value of a process, $\mathbb{E}$ , and the variance of that process, $Var$.

We will also be using the random walk as our canonical example, so please keep it in mind.

The probability of a fair coin landing heads is $ \frac 1 2 $ as is the probability for tails. We write this like

$ \mathbb{P}(H) = \mathbb{P} (T) = \frac 1 2$ 

To construct the random walk, we have to toss the coin several times, we will record the outcome of this coin tossing as $\omega$, such that if we toss a coin three times, with the values (H, T, T), then $\omega = \{ \omega_1 = H, \omega_2 = T, \omega_3 = T \}$. We then define the value of the random walk after $n$ steps:

$$M_n = \sum\limits_{j=1}^n X_j$$

where $X_j =  
\begin{cases} 
1 & if \hspace{2mm} \omega_j = H \\ 
-1 & if \hspace{2mm} \omega_j = T
\end{cases}~~~~$    and  $~~M_0 = 0$.

After one toss $M_1$ is either 1 or -1. Just to get some practice, let's calculate the probability that $M_1 = 1$ explicitly:

$\mathbb{P}(M_1 = 1) =\mathbb{P}(X_1 = 1) = \mathbb{P}(\omega_1 = H) = \frac 1 2$

After two tosses $M_2$ is either, 2, 0, or -2. Let's calculate all three probabilities explicitly:

$\mathbb{P}(M_2 = 2) =\mathbb{P}(X_1 = 1 ~~and ~~ X_2 = 1) = \frac 1 2 * \frac 1 2 = \frac 1 4$

$\mathbb{P}(M_2 = -2) =\mathbb{P}(X_1 = -1 ~~and ~~ X_2 = -1) = \frac 1 2 * \frac 1 2 = \frac 1 4$

$\mathbb{P}(M_2 = 0) =\mathbb{P}((X_1 = 1 ~~and ~~ X_2 = -1) ~~ or ~~(X_1 = -1 ~~and ~~ X_2 = 1)) = \frac 1 4 + \frac 1 4= \frac 1 2$


One thing that we are very interested in is the 'average value' of a process. While there are many ways to calculate the 'average value', the most useful definition in a theoretical setting is the 'mean' or expectation value. The definition of the expectation is simply the sum of all possible outcomes multiplied by their probabilities:

$\mathbb{E}(X) = \sum\limits_{\omega \in \Omega} X(\omega) \mathbb{P}(\omega) ~~~~ $  where  $\Omega$ is the set of all possible outcomes.

So, for the random walk, after one toss, we have

$\mathbb{E}(M_1) = 1 * \frac 1 2 + -1 * \frac 1 2 = 0$

And for two tosses, we have

$\mathbb{E}(M_2) = 2 * \frac 1 4 + 0 * \frac 1 2 -2 * \frac 1 4 = 0$

It is fairly easy to show that the random walk always has a special property that the expectation after $n$ tosses is always equal to the starting value:

$\mathbb{E}(M_n) = M_0 = 0$

Finally, let us examine the variance of a process. The variance of a process is a rather theoretical construct, but one can think of it as measuring how far the process on average varies from the expectation value of that process. To make it concrete, here is one definition:

$Var(X) = \mathbb{E} \left[(X - \mathbb{E}(X))^2 \right]$

One thing to point out, is that this is not the only definition of variance, however discussing this will take us too far from our present purpose. For those who are interested you can check out the [wikipedia article](http://en.wikipedia.org/wiki/Variance) for more information.

Let's calculate the variance for the random walk:

$Var(M_n) = \sum\limits_{j=1}^n Var(X_j) = \sum\limits_{j=1}^n \mathbb{E} \left[(\pm 1)^2 \right] = n$

#### Scaled Random Walk

It's a bit of a chicken and egg situation, but to make a good model of the price of a stock, we must introduce the scaled random walk: 

$W^{(n)}(t) = \frac 1 {\sqrt n} M_{nt} $

where we choose $nt$ to be an integer. 

The scaled random walk has the nice property that the expectation is zero, and the variance is one.

Let's program this up:

```python
seed(123)

price = [85]
for t in range(100):
    data = [0]
    for n in range(500):
        coin = randn()
        if coin<0:
            coin = -1
        else:
            coin = +1
        data.append(coin+data[n])
    price.append(price[t]+data[-1]/sqrt(500)) 
plot(price); show()    
```

As one can see this generates a very convincing model of the price of a stock.

The procedure can naturally be made better by letting $n$ (the number of steps per unit time) grow large. In fact, when $n \rightarrow \infty$ the scaled random walk gets a new name: Brownian Motion.
