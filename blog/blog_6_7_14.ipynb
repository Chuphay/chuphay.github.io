## Geometric Brownian Motion & a Little Bit of Stochastic Calculus

In my last blog post, we had introduced Brownian Motion as a better approximation to the price of a stock than a simple random walk. In this blog post we will introduce an even better model.

While Brownian Motion superficially seemed to replicate the price of a stock, there are some serious problems with that model. The first, and perhaps, most serious problem was that Brownian Motion can easily become negative. Obviously the price of a stock can never become negative. The worst it can do is drop to zero if the company goes bankrupt.

Black, Scholes and Merton invented a much better model in the early 1970s, for which they were awarded the Nobel prize in 1997. 

The basics of Geometric Brownian Motion are simple to understand. At each discrete time step, the price of a stock can either go up by an up factor $u=1+\alpha$ or down by a down factor $d=1-\alpha$. Distantly following Shreve's Stochastic Calculus for Finance, we will determine the price of stock at time $t$ by recording the result of $nt$ coin tosses, where $n$ is the number of subdivisions we take. If, at time $0$, the stock is priced at $S(0)$ then we expect

$S(t) = S(0)*(1+\sigma/\sqrt n )^{\# ~~ of~~ heads}*(1-\sigma/ \sqrt n)^{\#~~of ~~ tails}$

Let's program this up, to see what it looks like:

```python
sigma = 0.5
initial_price = 50
seed(123)

data = [initial_price]
heads = 0
for i in range(10000):
    flip = randn()
    if flip>0:
        data.append(data[i]*(1+sigma/100))
        heads += 1
    else:
        data.append(data[i]*(1-sigma/100))
    
plot(data)
print round(data[-1],4) == round(initial_price *((1+sigma/100)**(heads))*((1-sigma/100)**(10000-heads)),4)
```

It should not surprise anyone that as we take the limit $n \rightarrow \infty$ we get a simple closed form equation:

$S(t) = S(0)\exp\left[\sigma W(t) - \frac 1 2 \sigma^2 t\right]$

where $W(t)$ is the scaled random walk that we discussed in the previous blog post.

It is straightforward to prove this assertion using the mathematics already available to us, but I would rather take a different approach following Belal Baaquie's Quantum Finance that will introduce some new mathematics to us.

#### Stochastic Differential Equations

It is common in theoretical finance to model the price of a stock as a dfferential equation:

$$\frac {dS(t)} {dt} = \sigma S R(t)$$

where $\sigma$ is the same sigma as before and $R(t)$ is a Gaussian white noise closely related to the scaled random walk of the previous post. (we will make the distinction clear momentarily)

After discretizing time, $t = n \epsilon$, the white noise $R_n$ has the following probability distribution:

$$\sqrt{\frac {\epsilon}{2\pi}} exp{(-\frac {\epsilon} 2 R_n^2)} $$

The scaled walk, on the other hand, converges to the normal distribution with mean zero and variance $t$. i.e., the normal distribution has the probability distribution

$$\sqrt{\frac {1}{2\pi t}} exp{(-\frac {x^2} {2t})} $$

Thus,

$$Var(W^{(n)}(t)) = t$$

while

$$Var(R_n) = \frac 1 {\epsilon}$$

In other words, Gaussian white noise is simply the scaled random walk divided by the square root of the total time.

To perform th calculations needed to solve the differential equation above let us examine the derivative of an arbitrary function $f$ of the stock price $S(t)$:

$$\frac {df}{dt} = \lim_{\epsilon \rightarrow 0} \frac {f(t+\epsilon,S(t+\epsilon))-f(t,S(t))}{\epsilon}$$

Performing a Taylor expansion on the left hand side, we have:

$$\lim_{\epsilon \rightarrow 0 } \frac 1 {\epsilon} \left[f(t,S(t)) - f(t,S(t)) + \epsilon \frac {\partial f}{\partial t} + \epsilon \frac {\partial f}{\partial S} \frac {dS}{dt} + \frac {\epsilon^2}{2!} \frac {\partial^2 f}{\partial S^2} \left(\frac {dS}{dt}\right)^2 + ...\right]$$

Thus,

$$\frac {df}{dt} = \frac {\partial f}{\partial t} + \frac {\partial f}{\partial S} \frac {dS}{dt} + \lim_{\epsilon \rightarrow 0} \frac {\epsilon}{2!} \frac {\partial^2 f}{\partial S^2} \left(\frac {dS}{dt}\right)^2 +...$$

Now clearly, in ordinary calculus that last term would be zero, but as we have seen, the function $S(t)$ depends on white noise, so we have the following:

$$\left(\frac {dS}{dt}\right)^2 = \mathbb {E} \left( \sigma^2 S^2 R^2 \right) = \frac 1 {\epsilon} \sigma^2 S^2 $$ 

Which leads to the following differential equation:

$$\frac {df}{dt} = \frac {\partial f}{\partial t} + \sigma S \frac {\partial f}{\partial S} R + \frac {1}{2}\sigma^2 S^2\frac {\partial^2 f}{\partial S^2} $$

Now we are ready to turn back to our original differential equation. First, we define 

$$x(t) = \ln S(t)$$

So that,

$$ \frac {dx}{dt} = \sigma R(t) - \frac {\sigma^2} 2$$

Integrating yields

$$x(t) = x(0) - \frac {\sigma^2} 2 t + \sigma \int_0^t dt' R(t')$$

integrating over brownian motion can be easily approximated as I do in the following code:


```python
seed(123)
data = [0]
for i in range(100000):
    data.append(data[i]+randn())
plot(data)

integral = []

def integrate(data, grain):
    if grain>len(data):
        print "error"
        
    size = len(data)/grain
    #print size 
    out = [0]
    for i in xrange(1,size):
        out.append(out[0]+data[grain*i])
    return out    
                                     
figsize (25,20)        
subplot(221)
plot(integrate(data,10000))    
subplot(222)
plot(integrate(data,1000))  
subplot(223)
plot(integrate(data,100))   
subplot(224)
plot(integrate(data,10))    
show()
```

As I hope is apparent from looking at the above plots, integration of the Gaussian white noise leads to the scaled random walk, and thus

$S(t) = S(0)\exp\left[\sigma W(t) - \frac 1 2 \sigma^2 t\right]$
